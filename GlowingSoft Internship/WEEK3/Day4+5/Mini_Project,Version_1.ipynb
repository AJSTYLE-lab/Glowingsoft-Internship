#Web Scraping Of Restaurents From 'www.yelp.com':

First Getting Links of  5 cities of USA:

```python
url='https://www.yelp.com/search?find_desc=Delivery&find_loc=San%20Francisco%2C%20CA'#url only for San Francisco,CA
locations = {
    "Los Angeles, CA": "Los+Angeles%2C+CA%2C+United+States",
    "New York, NY": "New+York%2C+NY%2C+United+States",
    "San Francisco, CA": "San+Francisco%2C+CA%2C+United+States",
    "Florida": "Florida%2C+United+States",
    "California": "California%2C+United+States"
}

base_url = "https://www.yelp.com/search?find_desc=Delivery&find_loc="
generated_links = []
for location, query_param in locations.items():
    link = base_url + query_param
    generated_links.append(link)
    print(link)
```

Storing Responses Info As well:

As we are getting a response with a status code of 200 indicates a successful HTTP request.

```python
responses=[]
for url in generated_links:
    response = requests.get(url)
    responses.append(response)
responses
```

Extracting 5 html_codes of these Locations :

```python
html_codes=[]
for i in responses:
  html_code=BeautifulSoup(i.content,'html.parser')
  html_codes.append(html_code)
html_codes
```

Storing location and restaurant names together in a corresponding manner in .csv form:

```python
html_codes = []
for i in responses:
    html_code = BeautifulSoup(i.content, 'html.parser')
    html_codes.append(html_code)
Locations = []
Restaurants = {}
for i in range(len(html_codes)):
    location_name = html_codes[i].find('a', class_='css-19v1rkv')
    if location_name is not None:
        location = location_name.text.strip().replace(', CA', '').replace(', MD', '')
        Locations.append(location)
        menu_html_code = html_codes[i].find_all('a', class_='css-19v1rkv')
        restaurant_names = [item.text.strip() for item in menu_html_code[3:]]
        Restaurants[location] = restaurant_names
for location in Locations:
    csv_filename = f"{location}_restaurants.csv"
    with open(csv_filename, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(["Restaurant"])
        for restaurant_name in Restaurants[location]:
            writer.writerow([restaurant_name])
    print(f"Data for {location} saved to {csv_filename}")
```
