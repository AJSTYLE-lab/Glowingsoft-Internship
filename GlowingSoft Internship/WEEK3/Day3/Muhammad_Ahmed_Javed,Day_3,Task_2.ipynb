```python
!pip install requests
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
```

```python
url='https://www.boulevard-friedrichstrasse.de/speisen/'
url
```

```python
response=requests.get(url)
object=BeautifulSoup(response.content,'html.parser')
object
```

#Restaurent Name:

```python
restaurent_name='Restaurant_Boulevard-Friedrichstrasse'
```

```python
menu_tables = object.find_all('table')
item_name=[]
item_price=[]
for menu_table in menu_tables:
    menu_items = menu_table.find_all('tr')
    for menu_item in menu_items:
        name = menu_item.find('td', class_='kl')
        price = menu_item.find('td', class_='kr')
        if name and price:
            name_text = name.text.strip()
            price_text = price.text.strip()
            item_name.append(name_text)
            item_price.append(price_text)
            print("Item:",name_text)
            print("Price:",price_text)
            print('\n')
```

```python
print(item_name)
print(item_price)
```

```python
file=pd.DataFrame({'Item_Name':item_name,"Item_Price":item_price})
file.reset_index(drop=True,inplace=True)
```

```python
file
```

```python
file.to_excel(f'{restaurent_name}.xlsx',index=False)
```
