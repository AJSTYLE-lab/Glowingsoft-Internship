#How has retail economic activity in the United States changed over the past five years?

```python
List = []
for i in file.columns:
    if i.startswith("data.sales"):
        List.append(i)
yearlySales = file.groupby('time.year')[List].sum()
plt.title("Retail economic activity in the United States changed over the past five years")
plt.xlabel("Years 2012---->2017")
plt.ylabel("Sales Count")
plt.plot(yearlySales.index[20:], yearlySales.values[20:])
```

#What are the key differences between the Advance Monthly Retail Trade Survey (MARTS) and the Annual Retail Trade Survey (ARTS)?

```python
monthlysales = file.groupby('time.month')[List].sum()
plt.title("Advance Monthly Retail Trade Survey")
plt.xlabel("On Monthly ")
plt.ylabel("Sales Count")
plt.plot(monthlysales.index,monthlysales.values)
```
