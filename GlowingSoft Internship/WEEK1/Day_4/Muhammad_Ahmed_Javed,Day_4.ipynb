```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
url='https://40hadithnawawi.com/'
url
```

```python
response=requests.get(url)
response
```

```python
html_code=BeautifulSoup(response.content,'html.parser')
html_code
```

#TITLE OF .csv File

```python
title=html_code.find('title')
TITLE=title.text.strip('40HadithNawawi.com -')#TITLE will be the name of final file we will create.
```

#Finding 42 Hadith Numbers:

```python
hadith_numbers=html_code.find_all('div',class_='w-1/4 group-hover:bg-red-900 h-12 pl-3')
hadith_numbers
hadith_numbers_list=[]
for number in hadith_numbers:
  hadith_numbers_list.append(number.text)
hadith_numbers_list
```

#Finding the 42 Hadith titles:

```python
hadith_title_list=[]
hadith_titles=html_code.find_all('div',class_='w-3/4 group-hover:bg-red-900 h-auto pl-3')
for titles in hadith_titles:
  hadith_title_list.append(titles.text)
hadith_title_list
```

#Finding And Storing The Narrators:

generating Hadith Pages html Links:

```python
base_url = 'https://40hadithnawawi.com/hadith/'
hadith_links = []
responses=[]
for hadith_number in range(1, 43):
    hadith_url = f'{base_url}{hadith_number}-{hadith_title_list[hadith_number - 1].lower().replace(" ", "-")}'
    hadith_links.append(hadith_url)
hadith_links

```

Storing Resquests Responses:

```python
for i in hadith_links:
  response=requests.get(i)#requestinga all 42 hadith links and store it in responses(list)
  responses.append(response)
```

Now Storing Html_codes of these pages in list:

```python
html_codes_of_hadith_pages=[]
for i in responses:
  html_code=BeautifulSoup(i.content,'html.parser')
  html_codes_of_hadith_pages.append(html_code)
```

Extracting Hadith In English:

```python
data = {'Hadith_Numbers': hadith_numbers_list,'Hadith_Titles': hadith_title_list}
column_one = pd.DataFrame(data)
column_one.to_csv(f'{TITLE}.csv')
```

```python
import pandas as pd
import re
hadith_in_english = []
hadith_in_arabic = []
pattern = r'\[([^\]]+)\]'
book_names = []
for i in range(0, 41):
    a = html_codes_of_hadith_pages[i].find_all('div', class_='w-full md:mr-3 md:w-1/2 content-block')
    arabic = html_codes_of_hadith_pages[i].find_all('div', class_='w-full text-xl text-right md:w-1/2 content-block')
    for english_text, arabic_text in zip(a, arabic):
        hadith_in_english.append(english_text.text.strip())
        hadith_in_arabic.append(arabic_text.text.strip())
        book_name = re.findall(pattern, english_text.text.strip())
        book_names.append(book_name)
hadith_in_english_df = pd.DataFrame(hadith_in_english, columns=["Hadith_In_English"])
hadith_in_arabic_df = pd.DataFrame(hadith_in_arabic, columns=["Hadith_In_Arabic"])
result_df = pd.concat([hadith_in_english_df, hadith_in_arabic_df], axis=1)
result_df.to_csv("Hadiths_Data.csv")
```
